# LINE GRAPH
**Matplotlib is a low level graph plotting library in python that serves as a visualization utility**

```python
import matplotlib.pyplot as plt
```

```python
# line graph

plt.figure(figsize= (12,6))
plt.plot(df['math score'][:50], label = 'Math score',marker ='o')
plt.plot(df['writing score'][:50], label = 'Writing score',marker ='s')
plt.plot(df['reading score'][:50], label = 'Reading score',marker ='^')

plt.title("Performance")
plt.xlabel("Students")# x-axis
plt.ylabel("Score")#y-axis
plt.legend() # show the symbol to the line which describe the math,writing and reading score
plt.show()
```
